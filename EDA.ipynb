#Overview

Mercari Price Suggestion Challenge is a competition hosted on Kaggle. Mercari provides a  shopping app that enables sellers to list their items on the site for potential buyers.  Mercari tries to suggest prices for goods while the enlisting of goods takes place with the data provided by the users beforehand about the items. 

The site has images which are not part of the competition, whereas dat provided to us are the name , item condition ranging 1 to 5 from being good to worst, a category it belongs to, brand name, shipping with value 1 for free shipping or 0 for charges apply for shipping and item description. Lastly, there is the Price tag, the target value. 

The compitition uses root mean square log error metric.

There is one more dimension to this challenge other than a normal kaggle challenge. kernel here have a fixed capacity of 16 gb ram and 1 gb of hard disk space with only four cores to work with. On top  of that the code must get executed in under 60 minutes. 

The first stage of challenge has comparable amounts of train and test data. The next stage involved a new test set which is internally accessed by kaggel that is 5 times bigger than any of the yet provided datasets, so optimization is the next  challenge where code should not go default on memory  with that huge amount of data being  processed.

```python
#this function takes categories and returns sub categories in them
def sub_cat(row):
  if not isinstance(row,float):
    split = row.split("/")
    len_split = len(split)

    if len_split==3:
      return split[0],split[1],split[2],"None","None"

    elif len_split==4:
      return split[0],split[1],split[2],split[3],"None"
    
    elif len_split==5:
      return split[0],split[1],split[2],split[3],split[4]
  else: 
    return"None", "None","None","None","None"
```

#### As stated in the name EDA brands are pretty prevenlent in the name column so trying to find out if rows with NAN value in  brand_name column have any brand mentioned in the name column and if it can be used to impute the brand values. Also percentage of NAN brands_name columns will be measured which have an associative brand name in Names coulumn.

```python
#generation names which have brand and which do not have brand too in them
mercari.name[[8,22,429]]
```

```python
# getting all the brands name but removing "nan" values
brand_names =  list(mercari['brand_name'].dropna().unique())
```

```python
# A funciton to check if name feature of a row has an element belonging to brand_name list in it
def brand_check(brand_names, name, return_brand = True): 

    for brand in brand_names:
      #removing # signs name cols and finding brands if there.
      if re.search(brand, re.sub('([#])|([^a-zA-Z])',' ', name )):
        if return_brand:
          print(brand)
        return True
     
    return False
```

```python
# removing brands whit single character in them
garbage = []
for brands in brand_names:
  if len(brands) ==1:
    print(brands)
    brand_names.remove(brands)
```

#### Checking if brand_check is functinal or not

```python
brand_check(brand_names,'Galaxy S7 Edge (Unlocked) 32GB' )
```

```python
brand_check(brand_names,'Girls Nike Pro shorts' )
```

```python
brand_check(brand_names,'control text no brands' )
```

#### We can see that for 10 rows the brand_check takes 1.59 sec approximately hence it is impossible to go through all the 'nan' brand rows(as there are 632,682 of them) so using bootstrap method and sampling 30000 'nan' brand rows then finding if they have a brand name in corrosponding name feature(which takes about 1.5 hr) and getting a bootstraped percentage values of possible rows which can be imputed.

```python
df_universe = mercari[mercari['brand_name'].isna()]
```

```python
# finding True or False for every row if they have anykind of brand name in  the Name column
%%time
_ = df_universe['name'][:10].apply(lambda name: brand_check(brand_names, name,return_brand= False))
```

```python
size = 3000
df_pop = df_universe.sample(size)
df_pop['brand_in_name'] = df_pop['name'].apply(lambda name: brand_check(brand_names, name,return_brand= False))
```

```python
#overall percentage of imputation possible in the sample of 3000 rows 
percent = ((sum(df_pop['brand_in_name'])/size)*100)
percent
```

```python
# a sub sample of 500 with replacements is taken for 10000 times and percentage values are calculated against how much imputation is possible
per =[]
for i in range(10000):
  t=sum(df_pop['brand_in_name'].sample(size, replace = True))
  per.append((t/size)*100)
```

```python
# the distribution of percentage after bootstrapping
sns.distplot(per)
```

```python
print(f' 95 percent confidence interval for imputation perceantage is ({round(np.percentile(per,2.5),2)}, {round(np.percentile(per,97.5),2)})')
```

```python
np.percentile(per,50)
```

##### seeing the above values of percentage for passible imputation we can sat that 39 percent of 'nan' brand values can be imputed using name feature as brand names are prevelent in name feature as seen in the EDA of name.

```python
# little preprocessing the texts in item_description
def decontracted(phrase):
  # specific
  phrase = re.sub(r"won't", "will not", phrase)
  phrase = re.sub(r"can\'t", "can not", phrase)
  # general
  phrase = re.sub(r"n\'t", " not", phrase)
  phrase = re.sub(r"\'re", " are", phrase)
  phrase = re.sub(r"\'s", " is", phrase)
  phrase = re.sub(r"\'d", " would", phrase)
  phrase = re.sub(r"\'ll", " will", phrase)
  phrase = re.sub(r"\'t", " not", phrase)
  phrase = re.sub(r"\'ve", " have", phrase)
  phrase = re.sub(r"\'m", " am", phrase)
  return phrase



preprocessed_reviews = []
stop_words = set(stopwords.words("english"))
# tqdm is for printing the status bar
for sentance in tqdm(mercari['item_description'].values):
  try:
    sentance = re.sub(r"http\S+", "", sentance)
    sentance = BeautifulSoup(sentance, 'lxml').get_text()
    sentance = decontracted(sentance)
    sentance = re.sub("\S*\d\S*", "", sentance).strip()
    sentance = re.sub('[^A-Za-z]+', ' ', sentance)
    # https://gist.github.com/sebleier/554280
    sentance = ' '.join(e.lower() for e in sentance.split() if e.lower() not in stop_words)
    preprocessed_reviews.append(sentance.strip())
  except:
    pass
```

```python
def ShowWordcloud(data, title = None):
  
    wordcloud = WordCloud(
        background_color='white',
        stopwords=set(stopwords.words("english")),
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    
    
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show() 
```

## bivariate analysis

```python
#defing a function to check for ascending trend in price wrt lowering item_conditon
def extract_ascend_trend( super_df,col_name, group_by = 'item_condition_id', print_values = True, return_values =True, verbose = True):
  values = []
  #selecting this threshold with an intuition from extensive EDA
  if col_name == 'brand_name':
    thres = 0.10
  elif col_name == 'sub_cat1':
    thres = 0.12
  elif col_name == 'category_name':
    thres = 0.17
  #defining a function which takes a list and returns its prevelent Trend, False for not ascending and True for ascending
  def asc_trend(lst,threshold):
      bools = []
      cond_prev = lst[0]#setting the first ele here
      cond_1 = max(lst)#setting the max price form all the condition available 
      #if len of list is less than 2 (i.e items sold in only 1 condition out of 5) return Flase
      if len(lst)< 2 or [round(l) for l in lst].count(0)>3:
        return False
      
      for i,cond_curr in enumerate(lst[1:]):
        #if prev ele is greater then not ascending so bools.append(0)
        if cond_prev >= cond_curr:
          bools.append(0)
          
        else:
         #if current ele is greater but if the change is major(greater than threshold) wrt 
         # the prev price or the price in condition one better than the current
         #then only the ascending is true and bools.appen(1) else if minor change in price then
         #bools.append(0)  
          try:
            if (abs(cond_curr-cond_prev)/cond_prev) > threshold:
              bools.append(1)
            else:
              bools.append(0)
          except ZeroDivisionError:
            #try using the ele previous to the cond_prev
            try:
              if (abs(cond_curr-cond_prev)/lst[i-1]) > threshold:
                bools.append(1)
              else:
                bools.append(0)
            
            except IndexError:
              #boundry condition
              pass

        cond_prev = cond_curr
        
      #if 1's are more than 0's then a relatively ascending trend  
      if bools.count(1) > bools.count(0):
        return True
      else:
        return False
  ################## acs_trend ends here ##################################

  #for all the ele in a col_name    
  for i,value in enumerate(super_df[col_name].unique()):
    # for printing stuff
    if verbose :
      if i%2 == 0:
       print("\r checking for {}th element {}".format(i,value), end ="")
    #a df with just the iterated ele in col_name
    temp =  super_df[(super_df[col_name]== value)]
    #for that ele df checking if the trend is asc or not
    if asc_trend(temp.groupby(group_by)['price'].mean().to_list(),thres):
      values.append(value)
                                 
  if verbose :
    print("\r done\n", end="")  
  
  #for debugging purposes
  if print_values:
      print(values)
  #ret statement
  if return_values:
    return values
```
